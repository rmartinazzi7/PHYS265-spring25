```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def poten (xm, ym, x1, y1, m):
    p = -6.67E-11*m/np.sqrt((xm-x1)**2 +(ym-y1)**2)
    return p



fig, ax = plt.subplots(figsize=(9, 9))
x1val = np.linspace(1, 384400, 200)
y = np.linspace(0.0001, 12, 200)
y1 = np.linspace(0.0001, 12, 200)

earth = [poten(0, 0, x1, 0, 5.97E24) for x1 in x1val]

ax.plot(earth, y)
ax.set_yscale('log')

ax.set_xlim(min(earth), max(earth))  
ax.set_ylim(min(y), max(y)) 

ax.set_xlabel('Distance')
ax.set_ylabel('Gravitational Potential')


x_range = np.linspace(-1.5, 1.5, 100)
y_range = np.linspace(-1.5, 1.5, 100)

x_grid, y_grid = np.meshgrid(x_range, y_range)

z= poten(0,0, x_grid, y_grid, 5.97E24)


fig_3d, ax_3d_wire = plt.subplots(1,1,subplot_kw={"projection": "3d"})

ax_3d_wire.plot_wireframe(x_grid, y_grid, z, rstride= 10, cstride=10)


import matplotlib.cm as cm

ax_3d_wire.plot_surface(x_grid, y_grid, z, rstride= 10, cstride=10, cmap=cm.spring)
m= ax_3d_wire.plot_surface(x_grid, y_grid, z, rstride=10, cstride=10, cmap=cm.spring)
ax_3d_wire.view_init(elev=10, azim=-80)

fig_3d.colorbar(m,ax=ax_3d_wire)


```
